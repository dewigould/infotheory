```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
N = 100
uni_dist = [float(i) for i in range(N)]
def binom_dist(p):
    return [float(np.random.binomial(N,p)) for i in range(N)]
def pois_dist(p):
    return [float(np.random.poisson(p*N)) for i in range(N)]
```

```python
def KL_div(list1,list2):
    for i in list2:
        if i == 0.0:
            return np.inf
    result = 0
    for i in range(len(list1)):
        if list1[i] == 0.0:
            result += 0.0
        else:
            result -= list1[i]*math.log(list1[i]/list2[i],2.0)
    return result
```

```python
prob = np.linspace(0.01,1,25)
values = []
for i in prob:
    values.append(KL_div(uni_dist,binom_dist(i)))
```

```python
plt.figure()
plt.plot(prob,values)
plt.xlabel('p')
plt.ylabel('KL Divergence')
plt.show()
```

```python
values = []
for i in prob:
    values.append(KL_div(pois_dist(i),binom_dist(i)))
```

```python
plt.figure()
plt.plot(prob,values)
plt.xlabel('p')
plt.ylabel('KL Divergence')
plt.show()
```
